```python
import requests
from pydub import AudioSegment
import speech_recognition as sr
```

```python

# URL of the audio file
url = "https://github.com/vineetkukreti/vkaudio/raw/main/modi_news.mp3"


```

```python


# Download the audio file
response = requests.get(url)
with open("downloaded_audio.mp3", "wb") as f:
    f.write(response.content)

# Convert MP3 to WAV
audio = AudioSegment.from_mp3("downloaded_audio.mp3")
audio.export("converted_audio.wav", format="wav")


```

```python
# Transcribe the converted WAV audio
filename = "converted_audio.wav"
r = sr.Recognizer()
with sr.AudioFile(filename) as source:
    audio_data = r.record(source)
    text = r.recognize_google(audio_data)

print(text)

```
